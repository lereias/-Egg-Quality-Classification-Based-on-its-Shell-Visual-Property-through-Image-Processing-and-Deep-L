## Crop Image to Reduce Background

We shall crop the image in order to reduce the background so that the egg in the image would be more focused on. The cropping procedure will not affect the size diference of the eggs from one another.

```python

tadirect = directory + "/FourAngleEggs"
#crop width and height
width = 3600
height = 3600

for x in os.listdir(tadirect):
    directory2 = ""
    directory2 = tadirect + "/" + x     #images to be cropped
    
    curdir = os.listdir(directory2)
    
    for y in curdir:
        if x == "Balut-penoy":
            tempim1 = directory2 + "/" + y

            im1 = cv2.imread(tempim1)

            center = (im1.shape[0] / 2, im1.shape[1] / 2)
            #print(center)
            w = center[0] - width/2
            h = center[1] - height/2

            #slicing to perform cropping
            print("cropping " + y + " in dimensions (" + str(width) + "," + str(height) + ")")
            crop_img = im1[int(h):int(h+width), int(w):int(w+height)]

            #save and replace the image
            cv2.imwrite(tempim1, crop_img)
            print("crop finish " + y)
            
        else:
            for z in os.listdir(directory2 + "/" + y):
                tempim1 = directory2 + "/" + y + "/" + z

                im1 = cv2.imread(tempim1)

                center = (im1.shape[0] / 2, im1.shape[1] / 2)
                #print(center)
                w = center[0] - width/2
                h = center[1] - height/2

                #slicing to perform cropping
                print("cropping " + z + " in dimensions (" + str(width) + "," + str(height) + ")")
                crop_img = im1[int(h):int(h+width), int(w):int(w+height)]

                #save and replace the image
                cv2.imwrite(tempim1, crop_img)
                print("crop finish " + z)

print("\nAll finished")
```

## Concatenate Four Angles of Egg

Four images of an egg are taken (the front, back, top, bottom). We will combine all of them to represent the entire egg.

```python
tadirect = directory + "/FourAngleEggs"
combfour = directory + "/CombinedFourAngleEggs"

if not os.path.exists(combfour):
    os.makedirs(combfour)

#check if there is the right amount of images (1 for each angle of egg)
for w in os.listdir(tadirect):
    directory2 = ""
    directory2 = tadirect + "/" + w
    
    if w == "Balut-penoy":
        curdir = os.listdir(directory2)

        #number of files in directory
        filecnt = len([name for name in curdir])

        if filecnt%4 == 1:
            print("ERROR (" + w + "): Incorrect number of images.")
            sys.exit()
            
    else:
        for v in os.listdir(directory2):
            directory3 = directory2 + "/" + v
            
            curdir = os.listdir(directory3)
            
            #number of files in directory
            filecnt = len([name for name in curdir])

            if filecnt%4 == 1:
                print("ERROR (" + w + "): Incorrect number of images.")
                sys.exit()

for x in os.listdir(tadirect):
    directory2 = ""
    directory2 = tadirect + "/" + x                             #where the two images of two angles of egg is stored
    directory3 = directory  + "/CombinedFourAngleEggs/" + x     #where to put the combined image of the angles
    
    if x == "Balut-penoy":
        curdir = os.listdir(directory2)
        #number of files in directory
        filecnt = len([name for name in curdir])
        
        i = 1
        for y in range(filecnt):

            if y%4 == 0:
                tempim1 = directory2 + "/" + curdir[y]
                im1 = cv2.imread(tempim1)
                tempim2 = directory2 + "/" + curdir[y+1]
                im2 = cv2.imread(tempim2)
                tempim3 = directory2 + "/" + curdir[y+2]
                im3 = cv2.imread(tempim3)
                tempim4 = directory2 + "/" + curdir[y+3]
                im4 = cv2.imread(tempim4)

                #resize for ResNet50
                width = 224
                height = 224

                dim = (width,height)

                im_v = cv2.vconcat([cv2.resize(im1, dim), cv2.resize(im2, dim)])
                im_v2 = cv2.vconcat([cv2.resize(im3, dim), cv2.resize(im4, dim)])
                im_v3 = cv2.hconcat([im_v, im_v2])
                # show the output image
                #cv2.imshow("image",im_v)
                #cv2.waitkey(0)

                #save image
                #create folder to store concatenated images if not existing
                newpath = directory3 + "2"
                if not os.path.exists(newpath):
                    os.makedirs(newpath)

                #combine_img = newpath + "/IMG_" + str(random.randint(0,99999999)) + "_" + str(random.randint(0,999999)) + ".jpg"
                combine_img = newpath + "/" + x + " " + str(i) + ".jpg"

                i += 1

                cv2.imwrite(combine_img,cv2.resize(im_v3, dim))

                print(curdir[y] + " , " + curdir[y+1] + " , " + curdir[y+2] + " , " + curdir[y+3] + " combined.")
                 
    else:
        for y in os.listdir(directory2):
            directory4 = ""
            directory4 = directory2 + "/" + y
            
            curdir = os.listdir(directory4)
            #number of files in directory
            filecnt = len([name for name in curdir])
            
            i = 1
            for z in range(filecnt):
                if z%4 == 0:
                    tempim1 = directory4 + "/" + curdir[z]
                    im1 = cv2.imread(tempim1)
                    tempim2 = directory4 + "/" + curdir[z+1]
                    im2 = cv2.imread(tempim2)
                    tempim3 = directory4 + "/" + curdir[z+2]
                    im3 = cv2.imread(tempim3)
                    tempim4 = directory4 + "/" + curdir[z+3]
                    im4 = cv2.imread(tempim4)

                    #resize for ResNet50
                    width = 224
                    height = 224

                    dim = (width,height)

                    im_v = cv2.vconcat([cv2.resize(im1, dim), cv2.resize(im2, dim)])
                    im_v2 = cv2.vconcat([cv2.resize(im3, dim), cv2.resize(im4, dim)])
                    im_v3 = cv2.hconcat([im_v, im_v2])
                    # show the output image
                    #cv2.imshow("image",im_v)
                    #cv2.waitkey(0)

                    #save image
                    #create folder to store concatenated images if not existing
                    newpath = directory3 + "2"
                    if not os.path.exists(newpath):
                        os.makedirs(newpath)

                    #combine_img = newpath + "/IMG_" + str(random.randint(0,99999999)) + "_" + str(random.randint(0,999999)) + ".jpg"
                    combine_img = newpath + "/" + x + "_" + y + " " + str(i) + ".jpg"

                    i += 1

                    cv2.imwrite(combine_img,cv2.resize(im_v3, dim))

                    print(curdir[z] + " , " + curdir[z+1] + " , " + curdir[z+2] + " , " + curdir[z+3] + " combined.")
        
print("All finished")
```
