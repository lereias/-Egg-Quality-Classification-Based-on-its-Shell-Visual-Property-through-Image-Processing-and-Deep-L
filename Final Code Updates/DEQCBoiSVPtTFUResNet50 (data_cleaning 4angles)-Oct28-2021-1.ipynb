## Crop Image to Reduce Background

We shall crop the image in order to reduce the background so that the egg in the image would be more focused on. The cropping procedure will not affect the size diference of the eggs from one another.

```python
augdirect = directory + "/AugFourAngleEggs"
#crop width and height
width = 3600
height = 3600

for a in os.listdir(augdirect):
    augdirect2 = augdirect + "/" + a
    
    if a == "Balut-penoy3":
        print()
    
    print(a)
    print()
    
    if a == "Balut-penoy3":
        #access egg folders
        for b in os.listdir(augdirect2):
            augdirect3 = augdirect2 + "/" + b

            #access angle folders
            for c in os.listdir(augdirect3):
                augdirect4 = augdirect3 + "/" + c
                
                #access images
                for d in os.listdir(augdirect4):
                    tempim1 = augdirect4 + "/" + d

                    im1 = cv2.imread(tempim1)

                    center = (im1.shape[0] / 2, im1.shape[1] / 2)
                    #print(center)
                    w = center[0] - width/2
                    h = center[1] - height/2

                    #slicing to perform cropping
                    print("cropping " + d + " in dimensions (" + str(width) + "," + str(height) + ")")
                    crop_img = im1[int(h):int(h+width), int(w):int(w+height)]

                    #save and replace the image
                    cv2.imwrite(tempim1, crop_img)
                    print("crop finish " + d)
                    
    else:
        #access subfolders
        for b in os.listdir(augdirect2):
            print()
            print(b)
            print()
            
            augdirect3 = augdirect2 + "/" + b
            
            #access egg folders
            for c in os.listdir(augdirect3):
                augdirect4 = augdirect3 + "/" + c
      
                #access angle folders
                for d in os.listdir(augdirect4):
                    augdirect5 = augdirect4 + "/" + d

                    #access images
                    for e in os.listdir(augdirect5):
                        tempim1 = augdirect5 + "/" + e

                        im1 = cv2.imread(tempim1)

                        center = (im1.shape[0] / 2, im1.shape[1] / 2)
                        #print(center)
                        w = center[0] - width/2
                        h = center[1] - height/2

                        #slicing to perform cropping
                        print("cropping " + e + " in dimensions (" + str(width) + "," + str(height) + ")")
                        crop_img = im1[int(h):int(h+width), int(w):int(w+height)]

                        #save and replace the image
                        cv2.imwrite(tempim1, crop_img)
                        print("crop finish " + e)

print("All finished")
```
